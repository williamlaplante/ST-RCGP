```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def plot(y_observed=10, sigma_likelihood=2): 
    # Given prior and likelihood parameters
    mu_prior = 0        # Prior mean
    sigma_prior = 3     # Prior standard deviation
    sigma_prior2 = sigma_prior**2  # Prior variance

    sigma_likelihood2 = sigma_likelihood**2  # Likelihood variance



    # Compute posterior parameters (Normal-Normal update)
    sigma_posterior2 = 1 / (1/sigma_prior2 + 1/sigma_likelihood2)  # Posterior variance
    sigma_posterior = np.sqrt(sigma_posterior2)  # Posterior standard deviation
    mu_posterior = sigma_posterior2 * (mu_prior/sigma_prior2 + y_observed/sigma_likelihood2)  # Posterior mean

    (mu_posterior, sigma_posterior)

    # Generate x values for plotting
    x = np.linspace(-10, 20, 500)

    # Compute the distributions
    prior = norm.pdf(x, mu_prior, sigma_prior)
    likelihood = norm.pdf(x, y_observed, sigma_likelihood)
    posterior = norm.pdf(x, mu_posterior, sigma_posterior)

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(x, prior, label=r"Prior: $\theta \sim N(0,2)$", linestyle="dashed")
    plt.plot(x, likelihood, label=r"Likelihood: $N(\theta,2)$", linestyle="dotted")
    plt.plot(x, posterior, label=f"Posterior: N(1.5, {sigma_posterior:.2f})", linewidth=2)
    plt.axvline(y_observed, color='red', linestyle=":", label="Observed y=3")

    plt.xlabel("Î¸")
    plt.ylabel("Density")
    plt.title("Bayesian Update: Normal Prior and Likelihood")
    plt.legend()
    plt.xlim(x[0], x[-1])
    plt.show()
```

```python
#skinny tails
plot(y_observed=10, sigma_likelihood=1)
```

```python
#Fat tails
plot(y_observed=10, sigma_likelihood=8)
```
