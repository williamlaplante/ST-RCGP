```python
import sys
sys.path.append("../")
import matplotlib.pyplot as plt
import numpy as np
import nibabel as nib
```

```python
nii_file = '../../data/sub-9001/ses-2/dwi/sub-9001_ses-2_dwi.nii'

# Load the NIfTI file
img = nib.load(nii_file)

# Access image data and header
data = img.get_fdata()
header = img.header

print("shape : ", data.shape) # 256 by 256 images, 62 slices, and 50 time steps
```

```python
slice = 10

num_t_steps = 5
fig, axs = plt.subplots(num_t_steps, num_t_steps, figsize=(10,10), sharex=True, sharey=True)

t = 0
for i in range(num_t_steps):
    for j in range(num_t_steps):
        im = data[:, :, slice, t]
        im = (im - im.min()) / (im.max() - im.min())
        axs[i,j].imshow(data[:, :, slice, t], cmap='gray')
        axs[i,j].set_title(t)
        t+=1

plt.tight_layout()
    
```
