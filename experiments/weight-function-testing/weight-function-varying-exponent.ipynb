```python
import sys
sys.path.append('../../')

import torch as tc
import matplotlib.pyplot as plt
import matplotlib

from rcgp.rcgp import IMQ_and_gradient

plt.rc('font', family='serif', size=18)
plt.rcParams['text.usetex'] = True
matplotlib.rcParams['lines.linewidth'] = 2.5
```

```python
ymin, ymax = -20, 20
Y = tc.linspace(ymin, ymax, 1000)
m = tc.zeros_like(Y)
```

```python
plt.figure(figsize=(7,4))

for frac, alpha in zip(["1", "1/2", "1/4", "1/8"], [1, 1/2, 1/4, 1/8]):
    imq_out, imq_grad_out = IMQ_and_gradient(Y=Y, m=m, beta=1., c=1., alpha=alpha)

    plt.plot(Y, imq_out, label=r"$\alpha={}$".format(frac))


plt.legend(fontsize=14, frameon=False)
plt.xlim(Y[0], Y[-1])
plt.xlabel(r"$y$")
plt.ylabel(r"$w_{\small \textrm{IMQ}} (\mathbf{x}_k, y)$", fontsize=20, labelpad=8)
```
